```python
import cv2
import mediapipe as mp
import time
import math

mpDraw = mp.solutions.drawing_utils
mpPose = mp.solutions.pose
pose = mpPose.Pose()

cap = cv2.VideoCapture('data/20230511_161902.mp4')
pTime = 0

# 새로운 해상도 설정
new_width = 504
new_height = 896

while True:
    success, img = cap.read() # img => BGR

    # 영상이 끝난 경우 종료
    if not success:
        break

    # 영상 해상도 조정
    img = cv2.resize(img, (new_width, new_height))

    # convert to RGB for compatibility with the mediapipe library
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = pose.process(imgRGB)

    if results.pose_landmarks:
        mpDraw.draw_landmarks(img, results.pose_landmarks, mpPose.POSE_CONNECTIONS)

        # 어깨(Shoulder)와 엉덩이(Hip) 간의 거리 계산
        shoulder_landmark = results.pose_landmarks.landmark[mpPose.PoseLandmark.LEFT_SHOULDER]
        hip_landmark = results.pose_landmarks.landmark[mpPose.PoseLandmark.LEFT_HIP]

        # 좌표 값 추출
        shoulder_x, shoulder_y = int(shoulder_landmark.x * img.shape[1]), int(shoulder_landmark.y * img.shape[0])
        hip_x, hip_y = int(hip_landmark.x * img.shape[1]), int(hip_landmark.y * img.shape[0])

        # 좌표 간 거리 계산
        distance = math.sqrt((shoulder_x - hip_x) ** 2 + (shoulder_y - hip_y) ** 2)
        print("Shoulder to Hip distance:", distance)

    cTime = time.time()
    fps = 1 / (cTime - pTime)
    pTime = cTime

    cv2.putText(img, str(int(fps)), (70, 50), cv2.FONT_HERSHEY_PLAIN, 3, (255, 0, 0), 3)
    cv2.imshow("Image", img)
    cv2.waitKey(1) # 1ms delay

# 사용한 리소스 해제
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp
import numpy as np
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_pose = mp.solutions.pose

# 이미지 파일의 경우 이것을 사용하세요.:
IMAGE_FILES = ['data/20230511_162328.jpg']
BG_COLOR = (192, 192, 192)  # 회색
with mp_pose.Pose(
        static_image_mode=True,
        model_complexity=2,
        enable_segmentation=True,
        min_detection_confidence=0.5) as pose:
    for idx, file in enumerate(IMAGE_FILES):
        image = cv2.imread(file)
        image_height, image_width, _ = image.shape
        # 처리 전 BGR 이미지를 RGB로 변환합니다.
        results = pose.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

        if not results.pose_landmarks:
            continue
        print(
            f'Nose coordinates: ('
            f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.NOSE].x * image_width}, '
            f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.NOSE].y * image_height})'
        )

        annotated_image = image.copy()
        # 이미지를 분할합니다.
        # 경계 주변의 분할을 개선하려면 "image"가 있는
        # "results.segmentation_mask"에 공동 양방향 필터를 적용하는 것이 좋습니다.
        condition = np.stack((results.segmentation_mask,) * 3, axis=-1) > 0.1
        bg_image = np.zeros(image.shape, dtype=np.uint8)
        bg_image[:] = BG_COLOR
        annotated_image = np.where(condition, annotated_image, bg_image)
        # 이미지 위에 포즈 랜드마크를 그립니다.
        mp_drawing.draw_landmarks(
            annotated_image,
            results.pose_landmarks,
            mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=mp_drawing_styles.get_default_pose_landmarks_style())
        cv2.imwrite('/tmp/annotated_image' +
                    str(idx) + '.png', annotated_image)
        # 포즈 월드 랜드마크를 그립니다.
        mp_drawing.plot_landmarks(
            results.pose_world_landmarks, mp_pose.POSE_CONNECTIONS)
```

```python
import cv2
import mediapipe as mp
mp_drawing = mp.solutions.drawing_utils
mp_objectron = mp.solutions.objectron

# 이미지 파일의 경우 이것을 사용하세요.:
IMAGE_FILES = ['data/20230511_162328.jpg']
with mp_objectron.Objectron(static_image_mode=True,
                            max_num_objects=5,
                            min_detection_confidence=0.5,
                            model_name='Shoe') as objectron:
  for idx, file in enumerate(IMAGE_FILES):
    image = cv2.imread(file)
    # BGR 이미지를 RGB로 변환 후 객체 인식 작업을 처리합니다.
    results = objectron.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

    # 박스 랜드마크를 그립니다.
    if not results.detected_objects:
      print(f'No box landmarks detected on {file}')
      continue
    print(f'Box landmarks of {file}:')
    annotated_image = image.copy()
    for detected_object in results.detected_objects:
      mp_drawing.draw_landmarks(
          annotated_image, detected_object.landmarks_2d, mp_objectron.BOX_CONNECTIONS)
      mp_drawing.draw_axis(annotated_image, detected_object.rotation,
                           detected_object.translation)
      cv2.imwrite('/tmp/annotated_image' + str(idx) + '.png', annotated_image)
```

```python
import cv2
import numpy as np

# 영상 파일 경로 설정
video_path = "data/20230511_161902.mp4"

# 영상 파일 열기
cap = cv2.VideoCapture(video_path)

# 영상의 세로 크기와 프레임 수 가져오기
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# 출력 영상의 크기 설정
output_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) // 2
output_height = frame_height // 2 + 100  # y축 추가를 위해 100 픽셀 추가

# 비디오 저장을 위한 객체 초기화
fourcc = cv2.VideoWriter_fourcc(*'XVID')
output_video = cv2.VideoWriter('output.mp4', fourcc, 30, (output_width, output_height))

# 영상 프레임 반복
while cap.isOpened():
    # 프레임 읽기
    success, frame = cap.read()
    if not success:
        break

    # y축 선을 그릴 이미지 생성
    y_axis_image = np.zeros((100, output_width, 3), dtype=np.uint8)  # y축 영역을 검은색으로 초기화

    # y축 선 그리기
    cv2.line(y_axis_image, (50, 0), (50, 100), (0, 0, 255), thickness=2)  # 빨간색 선, 시작점: (50, 0), 끝점: (50, 100)

    # y축 선을 영상 위에 합성
    combined_frame = np.vstack((y_axis_image, frame))  # y축 영상과 원본 영상 합성

    # 결과 비디오 파일에 프레임 저장
    output_video.write(combined_frame)

    # 화면에 결과 출력
    cv2.imshow('Video with Y-Axis', combined_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 사용한 리소스 해제
cap.release()
output_video.release()
cv2.destroyAllWindows()
```
